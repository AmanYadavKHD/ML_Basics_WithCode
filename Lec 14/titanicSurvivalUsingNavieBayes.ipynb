```python
import pandas as pd
```

```python
df = pd.read_csv('titanic.csv')
df.head()
```

```python
df = df.drop(['PassengerId','Name','SibSp','Parch','Ticket','Cabin','Embarked'],axis='columns')
df.head()
```

```python
target = df.Survived
inputs = df.drop(['Survived'],axis='columns')
inputs.head()
```

```python
target.head()
```

```python
dummies = pd.get_dummies(inputs.Sex,dtype=int)
dummies.head(3)
```

```python
inputs = pd.concat([inputs,dummies],axis='columns')
inputs.head(3)
```

```python
inputs = inputs.drop(['Sex','male'],axis='columns')
inputs.head(3)
```

```python
inputs.columns[inputs.isna().any()]
#It will check that are there any Nan values in the dataframe.
```

```python
inputs.Age = inputs.Age.fillna(inputs.Age.mean())
inputs.head()
```

Now we split the dataset in training and testing.

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(inputs,target,test_size=0.2)
```

```python
len(X_train)
```

```python
len(X_test)
```

```python
from sklearn.naive_bayes import GaussianNB
model = GaussianNB()
model.fit(X_train,y_train)
```

```python
model.score(X_test,y_test)
```

```python
X_test[0:10]
```

```python
y_test[0:10]
```

```python
model.predict(X_test[0:10])
```

```python
model.predict_proba(X_test[:10])
#It gives an array of containing probability of survival and not survival of a passenger.
```

Calculation of score using cross validation.

```python
ss=[]
from sklearn.model_selection import cross_val_score
ss.append(cross_val_score(GaussianNB(),X_train,y_train,cv=5))
```
