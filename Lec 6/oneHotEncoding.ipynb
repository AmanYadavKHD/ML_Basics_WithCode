```python
import pandas as pd
```

```python
df = pd.read_csv("homeprices.csv")
df
```

```python
dummies = pd.get_dummies(df.town,dtype=int)
dummies 
#Here i have created a dummy data structure to store dummy values for the town column of df dataframe.
```

```python
#in pandas there is concat method to join two dataframes.
merged = pd.concat([df,dummies],axis="columns")
merged
```

```python
#we need to drop the town column and one from the dummy columns also .
final = merged.drop(['town','west windsor'],axis='columns')
final
```

```python
#Now creating the linear regression model.
from sklearn import linear_model
reg = linear_model.LinearRegression()
```

```python
x= final.drop(['price'],axis='columns')
x
```

```python
y=final.price
y
```

```python
reg.fit(x,y)
```

```python
reg.predict([[2600,0,1]])
#This predict the price of 2600sq-ft area house in robinsville
```

```python
reg.predict([[2800,0,0]])
#This predict the price of 2600sq-ft area house in west windsor
```

```python
#To check the accuracy of our model we use
reg.score(x,y)
#more it is close to 1,more is our model accurate.
```

```python
#This model is 95% accurate.
```
